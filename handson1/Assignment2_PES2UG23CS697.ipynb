```python
from transformers import pipeline

qa_bot = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

print(" QA model loaded successfully!")
```

```python
context = """
The giant panda is a bear species native to China, mainly found in Sichuan, Shaanxi, and Gansu provinces.
Pandas are known for their black-and-white fur and their diet which consists mostly of bamboo.
Although pandas belong to the order Carnivora, they rely on bamboo for the majority of their nutrition.
Adult pandas can weigh between 70 and 120 kilograms.
Pandas spend around 10 to 16 hours a day eating bamboo and rest for the remaining time.
They have strong jaw muscles and large molars to crush bamboo.
The species is considered a conservation success due to habitat protection and breeding programs.
Pandas communicate through sounds such as bleats and honks, and also use scent marking to communicate.
"""

print("Context loaded successfully!")
print("\nCONTEXT:\n")
print(context)
```

```python
# demo1: ask a question
question = "What do pandas mostly eat?"

result = qa_bot(question=question, context=context)
result
```

```python
#demo2: ask multiple questions
questions = [
    "Where are giant pandas native to?",
    "What do pandas mostly eat?",
    "How much do adult pandas weigh?",
    "How many hours do pandas eat in a day?",
    "How do pandas communicate?",
    "Why are pandas considered a conservation success?"
]

for q in questions:
    ans = qa_bot(question=q, context=context)
    print("Q:", q)
    print("A:", ans["answer"])
    print("Confidence Score:", round(ans["score"], 4))
    print("-" * 70)
```

```python
# Demo 3: User Input Mode (Ask your own question)
user_q = input("Ask a question about pandas: ")
ans = qa_bot(question=user_q, context=context)

print("\nAnswer:", ans["answer"])
print("Confidence Score:", round(ans["score"], 4))
```

```python
def ask_question(q, ctx, threshold=0.2):
    res = qa_bot(question=q, context=ctx)
    if res["score"] < threshold:
        return {
            "question": q,
            "answer": "Sorry, answer not clearly found in given context.",
            "confidence": round(res["score"], 4)
        }
    return {
        "question": q,
        "answer": res["answer"],
        "confidence": round(res["score"], 4)
    }

test_questions = [
    "What do pandas mostly eat?",
    "What is the capital of India?"  # not in context
]

for tq in test_questions:
    out = ask_question(tq, context)
    print("Q:", out["question"])
    print("A:", out["answer"])
    print("Confidence:", out["confidence"])
    print("-" * 70)
```
