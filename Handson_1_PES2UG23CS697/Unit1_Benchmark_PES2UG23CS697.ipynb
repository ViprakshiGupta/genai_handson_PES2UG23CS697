```python
from transformers import pipeline
```

```python
prompt = "The future of Artificial Intelligence is"

models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}

for name, model_id in models.items():
    print("\n==============================")
    print(f"MODEL: {name} ({model_id})")
    try:
        gen = pipeline("text-generation", model=model_id)
        out = gen(prompt, max_length=40)
        print("Output:", out[0]["generated_text"])
    except Exception as e:
        print("FAILED:", e)
```

```python
sentence_bert = "The goal of Generative AI is to [MASK] new content."
sentence_roberta = "The goal of Generative AI is to <mask> new content."

# BERT
print("---- BERT Fill Mask ----")
bert_mask = pipeline("fill-mask", model="bert-base-uncased")
print(bert_mask(sentence_bert)[:3])

# RoBERTa
print("\n---- RoBERTa Fill Mask ----")
roberta_mask = pipeline("fill-mask", model="roberta-base")
print(roberta_mask(sentence_roberta)[:3])

# BART
print("\n---- BART Fill Mask ----")
bart_mask = pipeline("fill-mask", model="facebook/bart-base")
print(bart_mask(sentence_roberta)[:3])
```

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

for name, model_id in models.items():
    print("\n==============================")
    print(f"MODEL: {name} ({model_id})")
    try:
        qa = pipeline("question-answering", model=model_id)
        ans = qa(question=question, context=context)
        print("Answer:", ans)
    except Exception as e:
        print("FAILED:", e)
```

| Task | Model | Classification (Success/Failure) | Observation (What actually happened?) | Why did this happen? (Architectural Reason) |
| :--- | :--- | :--- | :--- | :--- |
| **Generation** | BERT | Failure | Output was gibberish/dots like: “The future of Artificial Intelligence is.....” | BERT is an **encoder-only** model; it is not trained for autoregressive next-token generation. |
|  | RoBERTa | Failure | Output did not continue; mostly repeated the prompt without meaningful generation. | RoBERTa is also **encoder-only**, optimized for understanding, not generation. |
|  | BART | Partial / Weak | Model showed warnings and did not generate strong continuation using `text-generation` pipeline. | BART is **encoder-decoder**, but `text-generation` pipeline is not ideal for it; BART is mainly designed for seq2seq tasks. |
| **Fill-Mask** | BERT | Success | Predicted correct tokens like “create”, “generate”. | BERT is trained using **Masked Language Modeling (MLM)**, so fill-mask is its strength. |
|  | RoBERTa | Success | Predicted correct words like “generate”, “create” with good scores. | RoBERTa is an improved BERT trained heavily on **MLM**, hence performs well. |
|  | BART | Partial | Predicted words but with low scores; weaker than BERT/RoBERTa. | BART is trained for denoising seq2seq, not pure MLM word masking. |
| **QA** | BERT | Partial | Answer extracted was incomplete (“as hallucinations,”) with low confidence score. | Base BERT is not fine-tuned for QA; QA requires task-specific fine-tuning (SQuAD). |
|  | RoBERTa | Partial | Answer was incomplete (“such as hallucinations, bias, and”) with low confidence. | Base RoBERTa also lacks QA fine-tuning; architecture supports QA only after fine-tuning. |
|  | BART | Success | Produced correct answer: “risks such as hallucinations, bias, and deepfakes” (even with timeout warning). | BART’s encoder-decoder can handle sequence tasks better; however QA performance still depends on fine-tuning. |
